```python
import pandas as pd

df_final = pd.read_csv("C:/NYU/24fall/Bloomberg Capstone/data/final_merged.csv", parse_dates=['date'], index_col='Unnamed: 0')
df_final
```

```python
df_final.drop_duplicates(inplace=True)
df_final.reset_index(drop=True, inplace=True)
df_final
```

```python
df_final.describe()
```

```python
import yfinance as yf
import datetime
ticker_list = df_final['tic'].unique()
for tick in ticker_list:
    df_slice = df_final[df_final['tic'] == tick]
    start = df_slice['date'].min()#.strftime('%Y-%m-%d')
    end = df_slice['date'].max()+datetime.timedelta(days=2) #.strftime('%Y-%m-%d')
    close_data = yf.download(tick, start=start, end=end)[['Close','Volume']]
    close_data.columns = ['Close_'+tick, 'Volume_'+tick]
    df_final = df_final.merge(close_data, left_on='date', right_index=True, how='left')

df_final
```

```python
yf.download('QRTEB', start='2015-10-20', end='2024-07-04')['Close']
```

```python
def match_ticker_with_close(x):
    tick = x['tic']
    return x['Close_'+tick]
def match_ticker_with_volume(x):
    tick = x['tic']
    return x['Volume_'+tick]
df_final['Close'] = df_final.apply(match_ticker_with_close, axis=1)
df_final['Volume'] = df_final.apply(match_ticker_with_volume, axis=1)
df_final
```
