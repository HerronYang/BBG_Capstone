```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
bbg_data = pd.read_excel('bbg_draft_final.xlsx', sheet_name='Sheet1')
bbg_data
```

```python
# get symbols
def get_symbols(s:str):
    return s.split(' ')[0]

symbols = bbg_data['Symbol'].apply(get_symbols)
symbols
```

```python
sentiment = pd.read_excel('bbg_draft_final.xlsx', sheet_name='Sheet2')
sentiment
```

```python
# download data from yf

subsample = 1
start_date = '2015-01-01'
end_date = '2024-10-04'

for i in range(subsample):
    merged_data = yf.download(symbols[i], start=start_date, end=end_date, interval='1d')
    sentiment_data = pd.Series(sentiment['Unnamed: '+str(i*2+1)], index=pd.to_datetime(sentiment[symbols[i]+'  US Equity']))
    print(merged_data)

    print(sentiment_data)
```
